```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
"""
1. Time plots
2. Seasonal plots
3. Box plots
"""

apt_data = pd.read_csv('NZWN.csv')
apt_data.head()
```

```python
#make date and time column the index of data frame
apt_data.set_index('valid', inplace=True)
apt_data.index = pd.to_datetime(apt_data.index)

#Create descriptive statistics for weather data at wellington airport
apt_data.describe()
```

```python
##Convert vsby to numeric ( account for M values )
apt_data['vsby'] = pd.to_numeric(apt_data['vsby'], errors='coerce')

##Create time plot
plt.style.use("seaborn-v0_8")
apt_data['vsby'].plot(title='Visibility (vsby) vs time')
plt.xlabel('Date')
plt.ylabel('Visibility (miles)')
plt.savefig('assets/Visibility (vsby) vs time.png')
plt.show()
```

```python
## pull out seasonal patterns for visibility
## pull out seasonal patterns for wind
apt_data['year'] = apt_data.index.year
apt_data['month'] = apt_data.index.month
apt_data.reset_index(inplace=True)
apt_data['week'] = apt_data['valid'].dt.isocalendar().week
apt_data['day'] = apt_data['valid'].dt.dayofweek
apt_data['hour'] = apt_data['valid'].dt.hour
apt_data['day_str'] = apt_data['valid'].dt.strftime('%j')
apt_data['year_month'] = apt_data['valid'].dt.strftime('%m')
apt_data.set_index('valid', inplace=True)
```

```python
apt_data.head()
```

```python
## Seasonal patterns visibility on a monthly basis

np.random.seed(42)
df_plot = apt_data[['month', 'year', 'vsby']].dropna().groupby(['month', 'year']).mean()[['vsby']].reset_index()
years = df_plot['year'].unique()
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'black']

##plot for monthly wind patterns
plt.figure(figsize=(16,12))
for i, y in enumerate(years):
    if i < len(colors):
        plt.plot('month', 'vsby', data=df_plot[df_plot['year'] == y], color=colors[i], label=y)
        if y == 2018:
            plt.text(df_plot.loc[df_plot.year==y, :].shape[0]+0.3, df_plot.loc[df_plot.year==y, 'vsby'][-1:].values[0], y, fontsize=12, color=colors[i])
        else:
            plt.text(df_plot.loc[df_plot.year==y, :].shape[0]+0.1, df_plot.loc[df_plot.year==y, 'vsby'][-1:].values[0], y, fontsize=12, color=colors[i])

plt.gca().set(ylabel= 'vsby', xlabel = 'Month')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Seasonal Plot - Monthly Speeds", fontsize=20)
plt.ylabel('Wind speed (knots)]')
plt.xlabel('Month')
plt.savefig('assets/Seasonal_plot.png', dpi=300)
plt.show()
```

```python
## Seasonal patterns for wind speeds on a weekly basis
np.random.seed(42)
df_plot = apt_data[['month', 'week', 'vsby']].dropna()
df_plot = df_plot.groupby(['week', 'month']).mean()[['vsby']].reset_index()
df_plot = df_plot.sort_values(by='week', ascending=True)

# Defining the color palette
months = sorted(df_plot['month'].unique())
colors = plt.colormaps.get_cmap('tab20')(np.linspace(0, 1, len(months)))

# plot figures for weekly wind speeds
plt.figure(figsize=(16, 12))
for i, month in enumerate(months):
    month_data = df_plot[df_plot['month'] == month]
    plt.plot(month_data['week'], month_data['vsby'], color=colors[i], label=f'Month {month}', marker='o')
    # Place text at the last actual data point
    if len(month_data) > 0:
        last_week = month_data['week'].iloc[-1]
        last_sknt = month_data['vsby'].iloc[-1]
        plt.text(last_week, last_sknt, f'{month}', fontsize=12, color=colors[i])

# Setting labels and titles
plt.xlabel('Week of Year', fontsize=14)
plt.ylabel('Wind Speed (knots)', fontsize=14)
plt.title('Seasonal Plot - Weekly Average Wind Speeds', fontsize=20)
plt.xticks(fontsize=12, alpha=0.7)
plt.yticks(fontsize=12, alpha=0.7)
plt.legend(title='Month', loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('assets/Seasonal_plot_weekly.png', dpi=300)
plt.show()
```

```python
## Box splot on visibility observations
## Box plot on wind speed observations
import seaborn as sns
plt.figure(figsize=(7,4))
sns.boxplot(data= apt_data, x='vsby')
plt.xlabel('Visibility (miles)')
plt.title(f'Boxplot-Visibility vs time')
```

```python
#Distribution of monthly visibility
apt_data['year'] = apt_data.index.year
apt_data['month'] = apt_data.index.month
apt_data['year_month'] = apt_data.index.strftime('%Y-%m')

#filter for 2023 and later
apt_plot = apt_data[apt_data['year'] >= 2023]
#plot boxplot
plt.figure(figsize=(12,6))
sns.boxplot(x='year_month', y='vsby', data=apt_plot)
plt.title('Distribution of monthly visibility in miles')
plt.xlabel('Year-Month', color='red')
plt.ylabel("Visibility ( miles )", color='red')
plt.xticks(rotation=45)
plt.show()
```

```python
#Distribution of visibility by hour
plt.figure(figsize=(10,6))
sns.boxplot(x=apt_data['hour'], y=apt_data['sknt'])
plt.title('Distribution of visibility by hour of day')
plt.xlabel('Hour of Day')
plt.ylabel('Visibility (miles)', color='red')
plt.show()
```
