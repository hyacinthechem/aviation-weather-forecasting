```python
import numpy as np
from notebooks.training import training_mlp
from notebooks.training import training_mlp_extended

# track test loss

test_loss = 0.0

mae_history = []
mse_history = []
import torch
def test_model(trained_model, x_test, y_test):

  # put model in evaluation mode
  trained_model.eval()
  with torch.no_grad():
      predictions = trained_model(x_test)
      print(predictions)
      mae = torch.mean(torch.abs(predictions - y_test))
      mse = torch.mean((predictions - y_test) ** 2)
      print(f'MAE: {mae.item()}')
      print(f'MSE: {mse.item()}')
```

```python
from notebooks.training.training_mlp import weather_nn, x_test_processed_wind, y_test_processed_wind

# test model on predicting wind speed
test_model(weather_nn, x_test_processed_wind, y_test_processed_wind)

# test model on predicting visibility
test_model(weather_nn, x_test_processed_wind, y_test_processed_wind)
```
